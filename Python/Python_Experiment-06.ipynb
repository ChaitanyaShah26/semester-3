**Python program to store all even numbers from an array i**

```python
import array as arr

a = arr.array("i",[])

n = int(input("Enter the length of an Array : "))

print("Enter elements of the array: ")
for i in range(n):
    a.append(int(input()))

for i in range(n):
    print(a[i],end = " ")

print("\nEven elements in Array are : ")
for i in range(len(a)):
    if a[i] % 2 == 0:
        print(a[i],end = " ")
```

**Program to find a pair with highest product from a given array of integers.**

```python
import array as arr

a = arr.array("i",[])

n = int(input("Enter the number of elements in an array: "))

for i in range(n):
    a.append(int(input()))
print(a)

x, y = a[0], a[1]
for i in range(n-1):
    for j in range(i+1,n):
        if x*y < a[j]*a[i]:
            x = a[i]
            y = a[j]
print ("The pair is is:",[x,y])
```

**Write a Python program that takes two digits m (row) and n (column) as input and generates a two-dimensional array. The element value in the i-th row and j-th column of the array should be i*j.
Note :
i = 0,1.., m-1
j = 0,1, n-1.**



```python
import array as arr

a=arr.array("i",[])

m = int(input("Enter the number of rows : "))
n = int(input("Enter the number of columns : "))

for i in range(m):
    for j in range(n):
        print(i*j,end="  ")
    print()
```
